```python
cv = {
    500: 0.722,
    750: 0.74,
    1000: 0.7635,
    1250: 0.786,
    1500: 0.8075,
    1750: 0.826,
    2000: 0.8425,
    2250: 0.856,
    2500: 0.87,
    2750: 0.8833,
    3000: 0.8925,
    3250: 0.895,
    3500: 0.8975
}

cp = {
    500: 1.01,
    750: 1.0255,
    1000: 1.05,
    1250: 1.073,
    1500: 1.095,
    1750: 1.113,
    2000: 1.134,
    2250: 1.143,
    2500: 1.1565,
    2750: 1.168,
    3000: 1.179,
    3250: 1.1825,
    3500: 1.185
}

def find_key(num):
    ans = 500
    while ans < num:
        ans += 250

    return ans


def interpolate_cv(x, upper, lower):
    k = (cv[upper] - cv[lower])/(upper - lower)
    m = cv[lower] - k * lower
    return round(k * x + m, 4)


def interpolate_cp(x, upper, lower):
    k = (cp[upper] - cp[lower]) / (upper - lower)
    m = cp[lower] - k * lower
    return round(k * x + m, 4)


def interpolate(choice, temp):
    if 500 <= temp <= 3500:
        if 500 < temp < 750:
            if choice == "cv":
                return interpolate_cv(temp, 750, 500)
            else:
                return interpolate_cp(temp, 750, 500)
        elif temp % 250 == 0:
            if choice == "cv":
                return cv[temp]
            else:
                return cp[temp]
        else:
            key = find_key(temp)
            if choice == "cv":
                return interpolate_cv(temp, key, key - 250)
            else:
                return interpolate_cp(temp, key, key - 250)
    else:
        print("\nTemp outside usable range!\nRange: 500K - 3500K")
        return -1

# interpolate("cp", 900)  <-- Exempel
```
